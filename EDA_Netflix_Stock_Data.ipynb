```python
# Importing
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Load dataset
df = pd.read_csv("Netflix_historical_data.csv")

# Convert Date column to datetime
df["Date"] = pd.to_datetime(df["Date"])
df
```

Datetime function call

```python
# The Datetime col is an obj, we need to convert it to Date type
df["Date"] = pd.to_datetime(df["Date"], utc=True, errors="coerce")
df = df.sort_values("Date").reset_index(drop=True)

print(df.dtypes)
print()
print(df[["Date"]].head())
```

Splicing

```python
# Splicing with loc and iloc
print("Splicing with .loc (first 10 rows)")
print(df.loc[:10, ["Date","Open","High","Low","Close","Volume"]])

print("\nSplicing with .iloc (rows 100–109)")
df.iloc[100:110, 0:6]
```

Having conditions using Mask



```python
print("Close > 500")
df[df["Close"] > 500].head()
```

```python
print("\nClose > 500 & Open < 500")
df[(df["Close"] > 500) & (df["Open"] < 500)].head()
```

```python
print("\nClose > 500 & Open < 500")
df[(df["Close"] > 500) & (df["Open"] < 500)].head()
```

```python
print("\nClose > 500 & Open < 500")
df[(df["Close"] > 500) & (df["Open"] < 500)].head()
```

```python
print("\nDays where Close > Open (up days)")
df[df["Close"] > df["Open"]].head()
```

```python
print("\nTop 1% volume days")
vol_thresh = df["Volume"].quantile(0.99)
print("Threshold:", vol_thresh)
df[df["Volume"] >= vol_thresh].head()
```

Filtering using query



```python
print("High volume AND price dropped (Close < Open)")
df.query("Volume >= @vol_thresh and Close < Open").head()
```

Sorting Normally, and Sorting with Lambda

```python
# Sorting Normally
print("Top 10 by Close")
df.sort_values(by="Close", ascending=False).head(10)
```

```python
# Sorting Using Lambda
print("Sorting by Close % 10")
df.sort_values(by="Close", key=lambda s: s % 10).head(10)[["Date","Close"]]
```

Aggregation

```python
# Group by year and get summary stats using aggregation
df["Year"] = df["Date"].dt.year
yearly_agg = df.groupby("Year")["Close"].agg(["count","mean","median","min","max"])
yearly_agg.head(10)
```

```python
# Group by month period and get mean and max using aggregation
monthly_agg = df.groupby(df["Date"].dt.to_period("M"))["Close"].agg(["mean","max"]).reset_index()
monthly_agg["Month"] = monthly_agg["Date"].astype(str)
monthly_agg.head(20)
```

Transform

```python
df["Close_norm"] = df["Close"].transform(lambda x: (x - x.min()) / (x.max() - x.min()))
df
```

Apply

```python
df["DailyReturn"] = df["Close"].pct_change()
print(df[["Date","Close","DailyReturn"]].head())

# Applying a custom function for rows, using isin
def day_label(row):
    if pd.isna(row["DailyReturn"]):
        return "Start"
    if row["DailyReturn"] > 0.02:
        return "Big Up"
    if row["DailyReturn"] < -0.02:
        return "Big Down"
    return "Normal"

df["DayLabel"] = df.apply(day_label, axis=1)
df[["Date","DailyReturn","DayLabel"]].head(20)
# print(df["DayLabel"].value_counts())
```

```python
# Find the day with largest range per year
# Using groupby and apply
df["Range"] = df["High"] - df["Low"]
top_range_per_year = (
    df.groupby("Year", group_keys=False)
      .apply(lambda g: g.nlargest(1, "Range")[["Date","Open","High","Low","Close","Range"]])
      .reset_index(drop=True)
)
top_range_per_year.head(20)
```

Visualization

```python
# Visualizing the closing price over time
plt.plot(df["Date"], df["Close"])
plt.title("Netflix Closing Price Over Time")
plt.xlabel("Date")
plt.ylabel("Close")
plt.show()
```

```python
# Visualizing the histogram of returns
plt.hist(df["DailyReturn"].dropna(), bins=50)
plt.title("Histogram of Daily Returns")
plt.xlabel("Daily Return")
plt.ylabel("Frequency")
plt.show()
```

```python
# Visualizing monthly mean closing
monthly_mean_close = df.groupby(df["Date"].dt.to_period("M"))["Close"].mean()
monthly_mean_close.plot()
plt.title("Monthly Mean Close (Resampled)")
plt.ylabel("Mean Close")
plt.show()
```

```python
# Visualizing the returns by weekday
df["Weekday"] = df["Date"].dt.day_name()
order = ["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]
data = [df.loc[df["Weekday"]==d, "DailyReturn"].dropna() for d in order]

plt.boxplot(data, labels=order, showfliers=False)
plt.title("Daily Returns by Weekday")
plt.ylabel("Daily Return")
plt.xticks(rotation=15)
plt.show()
```
